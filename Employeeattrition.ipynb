# Employee Attrition

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
#For loading data
df=pd.read_csv("Employee-Attrition.csv")
df


```

```python
#To get the info so that we can delete unwanted rows and columns
df.info()
```

```python
#Delete the column
df = df.drop(['EmployeeNumber','EnvironmentSatisfaction','HourlyRate','JobSatisfaction','MaritalStatus','MonthlyRate','Over18','OverTime','RelationshipSatisfaction',
            'StandardHours','StockOptionLevel','TrainingTimesLastYear','YearsAtCompany','EmployeeCount','YearsSinceLastPromotion','YearsWithCurrManager'], axis=1)
df
```

```python
#Droppping off the row so that we can focus on 1 to 50 clearly
df = df.drop(df.iloc[51:1471].index)
df
```

# PIE CHART

```python
#Pie Chart
#Data for Pie from Attrition
Attrition_rate= df['Attrition'].value_counts().to_dict() #We are adding the value counts to dict so that we get a count of yes and no
colors=("cyan","lightgreen")
sizes = Attrition_rate.values()
vlabels = ['Yes','No']

#Customizing Pie
plt.title('Attrition Rate',fontweight='bold',fontstyle='italic')
plt.axis('equal')
plt.pie(sizes, labels=vlabels, autopct='%1.1f%%',colors=colors)
plt.savefig("Pie_Attrition1.png")
plt.show()
```

# CORRELATION MATRIX

```python
numeric_columns  = df.select_dtypes(include=['number']) #Only takes numerical values from dataset
corr_matrix      =numeric_columns.corr()
#Customizing
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.1f',cbar=True,linewidth=0.5)
plt.title('Correlation Matrix',fontstyle='italic',fontweight='bold')
plt.savefig("correlation_matrix.png")
plt.show()
```

# Scatter Graph

```python
#Scatter Graph

plt.figure(figsize=(10,6),dpi=200)
plt.scatter(df['Age'] ,df['MonthlyIncome'],color='cyan',marker='o')
plt.title('Monthly Income By Age',fontstyle='italic',fontweight='bold')
plt.xlabel('Age',fontsize='10')
plt.ylabel('Monthly Income',fontsize='10')
plt.savefig("Income_byage.png")
plt.show()
```

```python
#Statistical Relation
from scipy.stats import skew
from scipy.stats import kurtosis

numeric_columns =df.select_dtypes(include=['number'])
kurtosis_values =numeric_columns.apply(kurtosis)
mean_values     =numeric_columns.mean()
std_values      =numeric_columns.std()
skewness_values =numeric_columns.skew()

Overall_stats = pd.DataFrame({'Mean' :mean_values,
                             'Standard Deviation':std_values,
                             'Skewness': skewness_values,
                             'Kurtosis Values' : kurtosis_values})

print(Overall_stats)
```
